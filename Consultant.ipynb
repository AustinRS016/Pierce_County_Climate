```python
import pandas as pd
import geopandas


df = pd.read_csv('carto_consult_climate_2010_2020.csv')

df = df.drop(columns=['FOG (Y/N)','SMOKE/HAZE (Y/N)'])

df


```

```python
df['year'] = pd.DatetimeIndex(df['DATE']).year

days = df.set_index(["NAME","DATE"]).count(level="NAME")
days = days.rename(columns={'STATION':'totalDays'})
days = days['totalDays'].reset_index()


days['fract_ofYear'] = days['totalDays']/365
days = days.set_index('NAME')

days
```

```python

# Group by Station and sum variables
x = df.groupby(['NAME'])[['SNOWFALL (INCHES)','PRECIPITATION (INCHES)','SNOW DEPTH (INCHES)']].sum().reset_index()
x = x.set_index('NAME')
x
```

```python
# Create df with coordinates and set name as index for concatenating with other dfs
latlng = df[['NAME','LATITUDE',"LONGITUDE","ELEVATION (METERS)"]]
latlng = latlng.drop_duplicates()
latlng = latlng.set_index('NAME')
latlng
```

```python
z = pd.concat([x,latlng,days], axis=1)

varis = ['SNOWFALL (INCHES)','PRECIPITATION (INCHES)']
#totalYears represents how many different years the station collected data in
#fract_ofYear represents the numerically exact amount of years the station was collecting data for (totalDays/365)

# Average snowfall and precipitation by unit of 1 year
for var in varis:
    z[var] = z[var]/z['fract_ofYear']   

#Average snow depth by day. Snow depth at 8 inches per day doesn't mean it's 8 inches deeper rather the
#snow depth hasn't changed 
z['SNOW DEPTH (INCHES)'] = z['SNOW DEPTH (INCHES)']/z['totalDays']

z = z.reset_index()
z


```

```python
#rename columns an df to be succinct add year.days column
final = z.rename(columns={'SNOW DEPTH (INCHES)': 'avgDailySnDepth(in)','PRECIPITATION (INCHES)':'avgYearlyPrecip(in)','SNOWFALL (INCHES)':'avgYearlySnowfall(in)'})

final
```

```python
final.to_csv('pierceWeatherStats.csv')
```

```python
gdf = geopandas.GeoDataFrame(final, geometry=geopandas.points_from_xy(final.LONGITUDE, final.LATITUDE))


gdf.to_file('pierceWeatherStats.json', driver='GeoJSON')
```
